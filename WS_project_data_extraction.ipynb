```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                  "AppleWebKit/537.36 (KHTML, like Gecko) "
                  "Chrome/122.0.0.0 Safari/537.36"
}
```

```python
test_data=[]
pages_to_scrape=64
for i in range (1,pages_to_scrape+1):
    print(f"Scraping page {i}...")
    
    url=f"https://stats.espncricinfo.com/ci/engine/stats/index.html?class=1;home_or_away=1;home_or_away=2;home_or_away=3;page={i};template=results;type=batting"
    response = requests.get(url, headers=headers)
    html_content = response.text
    soup = BeautifulSoup(html_content)
    soup1 = BeautifulSoup(html_content, "html.parser")
    table = soup1.find("div", class_="pnl650M")
    if table:
        rows = table.find_all("tr",class_="data1")
         
        for row in rows:
            cols = [td.text.strip() for td in row.find_all(["td", "th"])]
            test_data.append(cols)
        print(f"✅ Page {i} done — total rows: {len(test_data)}")
            
    else:
        print("error")

print(f"\n Scraping complete! Total records: {len(test_data)}")

# Test_df=pd.DataFrame(test_data)
# Test_df    
# header1=["Player","Span","Mat","Inns","NO","Runs","HS","Ave","100","50","0",""]	
# rows=test_data[:]
# Test_df=pd.DataFrame(rows,columns=header1)
# Test_df.to_csv(r'C:\Users\anura\OneDrive\Desktop\webscrapping\Project\scraped_data.csv', index=False)  
# Test_df 
header1=["Player","Span","Mat","Inns","NO","Runs","HS","Ave","100","50","0",""]	
rows=test_data[:]
Test_csv=pd.DataFrame(rows,columns=header1)
Test_csv.to_csv(r'C:\Users\anura\OneDrive\Desktop\webscrapping\Project\scraped_data.csv', index=False)  
Test_df=pd.read_csv(r"C:\Users\anura\OneDrive\Desktop\webscrapping\Project\scraped_data.csv") 
Test_df
```
